```python
import matplotlib.pyplot
import  numpy as np
import random as rd
```

```python
A=["A","B","O"]
EXP={
    ("A","A"):"E_A", 
    ("B","B"):"E_B", 
    ("A","B"):"E_AB", 
    ("B","A"):"E_AB", 
    ("O","O"):"E_O", 
    ("O","A"):"E_A",
    ("A","O"):"E_A", 
    ("O","B"):"E_B",
    ("B","O"):"E_B"
}

PROB={
    "E_A":0.2,
    "E_B":0.4,
    "E_AB":0.1,
    "E_O":0.3
}

CONST_POPULATION=100
Pop=[(rd.choice(A),rd.choice(A)) for _ in range (CONST_POPULATION)]
print (Pop)
```

```python
def EXP_POP(POP):
    L=[]
    for t in POP:
        L.append(EXP[t])
    return L

print (EXP_POP(Pop))
```
